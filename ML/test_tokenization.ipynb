```python
from transformers import BertTokenizer

# 1. Загружаем стандартный токенизатор BERT
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# 2. Наш химический "монстр"
chem_word = "Methylethylketone"

# 3. Смотрим, как его видит обычная модель
tokens = tokenizer.tokenize(chem_word)

print(f"Слово: {chem_word}")
print(f"Токены BERT: {tokens}")
```
